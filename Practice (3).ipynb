```python
def calculator():

  print("Welcome to the calculator")

  num1 = int(input("Enter First Number:  "))
  operators = (input("Enter an operator between + - * / :  "))
  num2 = int(input("Enter Second Number:  "))


  if operators == "+":
    print(num1 + num2)
  elif operators == "-":
    print(num1 - num2)
  elif operators == "*":
    print(num1 * num2)
  elif operators == "/":
    print(num1 / num2)
  else:
    print("Invalid operator")

calculator()
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))


largest = max(num1, num2, num3)

smallest = min(num1, num2, num3)

if num1 == num2 == num3:
  equality = "All Numbers are equal"
  print(equality)
else:
  print(f"The largest number is {largest}")
  print(f"The smallest number is {smallest}")
```

```python
fruits = []

for i in range(5):
  fruit = input("Enter a fruit: ")
  fruits.append(fruit)

print(fruits)

print("Your fruits: ", fruits)

print("The first fruit is: ", fruits[0])
print("The last fruit is: ", fruits[-1])

print("Reserved List: ", fruits[::-1])

print("Total Fruits: ", len(fruits))
```

```python
colors = set()

for i in range(5):
  color = input(f"Enter color {i+1}: ")
  colors.add(color)

print("Your color set", colors)

if "Red" in colors:
  print("Is red in the color set? Yes")

else:
  print("Is red in the color set? No")
```
